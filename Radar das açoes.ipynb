```python
!pip install yfinance
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
acoes = ['PETR4.sa','PETR3.sa']
```

```python
dados_acoes = yf.download(acoes,
                          period='1d',
                          start='2022-01-01')
```

```python
dados_acoes['Close'].plot()
plt.xlabel('data')
plt.ylabel('Preço do Fechamento')
plt.title('rastreador de Ações')
plt.legend(acoes, fontsize=7)
plt.show()
```

```python
#varias açoes para o mesmo ponto inicial para comprar a evolução
dados_normalizados = dados_acoes['Close'] / dados_acoes['Close'].iloc[0]
dados_normalizados.plot()
dados_acoes['Close'].plot()
plt.xlabel('data')
plt.ylabel('Preço do Fechamento')
plt.title('rastreador de Ações')
plt.legend(acoes, fontsize=7)
plt.show()
```
